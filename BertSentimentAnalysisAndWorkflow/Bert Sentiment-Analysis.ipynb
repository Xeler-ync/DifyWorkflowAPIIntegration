```python
# If your computer is also a Mac with Apple Silicon (M1/M2) and you want to install the Miniforge version of Transformers. 
# Next open a terminal and run code blow: curl -LO https://github.com/conda-forge/miniforge/releases/latest/download/Miniforge3-MacOSX-arm64.sh
# Then run: bash Miniforge3-MacOSX-arm64.sh
# Follow the instructions to complete the installation.
# After the installation is complete, close the terminal and open a new terminal.
# Then create a new conda environment with Python 3.9: conda create -n transformers python=3.9
# Activate the environment by runing conda activate transformers
# Finally, install the Transformers library with: pip install transformers
# conda create -n bert python=3.9
# conda activate bert
# pip install transformers
# pip install torch # For Mac M1/M2, you may need to install PyTorch following the instructions at https://pytorch.org/get-started/locally/
# If you are in a jupyter notebook, you may need to restart the kernel to make sure the newly installed packages are available.
# pip install ipykernel # python -m ipykernel install --user --name=bert --display-name="python-bert" 
# I use the model "uer/roberta-base-finetuned-jd-binary-chinese" from Hugging face model hub.
# pipeline for sentiment analysis in classifier and use the model to analyze the sentiment of a Chinese text.
# blow is the code to run sentiment analysis using a pre-trained model for Chinese text.

from transformers import pipeline


classifier = pipeline(
    "sentiment-analysis",
    model="uer/roberta-base-finetuned-jd-binary-chinese"
)

result = classifier("这家酒店服务真的很贴心，下次还会再来！")
print(result)
```
